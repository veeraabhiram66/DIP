```python
#importing libraries
import cv2 as cv
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv.imread('Lena.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Original Image')
plt.show()
#convert to grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Gray Image')
plt.show()
```

(A) Translation

```python
#applying translation on grayscale image
#translation wrt centre
#without using inbuilt functions
def translation(img, a, b):
    rows, cols = img.shape
    centre_x = rows//2
    centre_y = cols//2
    new_img = np.zeros((rows, cols), np.uint8)
    for i in range(rows):
        for j in range(cols):
            x = i - centre_x
            y = j - centre_y
            new_x = x + a
            new_y = y + b
            new_i = new_x + centre_x
            new_j = new_y + centre_y
            if new_i>=0 and new_i<rows and new_j>=0 and new_j<cols:
                new_img[new_i, new_j] = img[i, j]
    return new_img

translated_img = translation(gray, 100, 100)
plt.imshow(translated_img, cmap='gray')
plt.title('Translated Image')
plt.show()
```

(B) Scaling

```python
#applying scaling on grayscale image
#scaling wrt centre
#without using inbuilt functions
def scaling(img, a, b):
    rows, cols = img.shape
    new_rows = int(rows * a)
    new_cols = int(cols * b)
    new_img = np.zeros((new_rows, new_cols), dtype=np.uint8)
    center_row = rows // 2
    center_col = cols // 2
    new_center_row = new_rows // 2
    new_center_col = new_cols // 2
    
    for i in range(new_rows):
        for j in range(new_cols):
            orig_i = int((i - new_center_row) / a + center_row)
            orig_j = int((j - new_center_col) / b + center_col)
            if 0 <= orig_i < rows and 0 <= orig_j < cols:
                new_img[i, j] = img[orig_i, orig_j]
            
    return new_img
scaled_img_center = scaling(gray, 2, 2)
plt.imshow(scaled_img_center, cmap='gray')
plt.title('Scaled Image (Center)')
plt.show()
```

(C) Rotation

```python
#rotation of grayscale image
#rotation wrt centre
#without using inbuilt functions
def rotation(img, angle):
    rows, cols = img.shape
    new_img = np.zeros((rows, cols), dtype=np.uint8)
    center_row = rows // 2
    center_col = cols // 2
    angle = math.radians(angle)
    final_rows = round(rows * abs(math.cos(angle)) + cols * abs(math.sin(angle)))
    final_cols = round(rows * abs(math.sin(angle)) + cols * abs(math.cos(angle)))
    final_img = np.zeros((final_rows, final_cols), dtype=np.uint8)
    final_center_row = final_rows // 2
    final_center_col = final_cols // 2
    for i in  range(final_rows):
        for j in range(final_cols):
            orig_i = int((i - final_center_row) * math.cos(angle) + (j - final_center_col) * math.sin(angle) + center_row)
            orig_j = int(-(i - final_center_row) * math.sin(angle) + (j - final_center_col) * math.cos(angle) + center_col)
            if 0 <= orig_i < rows and 0 <= orig_j < cols:
                final_img[i, j] = img[orig_i, orig_j]
    return final_img

rotated_img = rotation(gray, 45)
plt.imshow(rotated_img, cmap='gray')
plt.title('Rotated Image')
plt.show()
```

(D) Shearing in X - direction

```python
#shearing in x-direction
#shearing wrt centre
#without using inbuilt functions
def shearing_x(img, a):
    rows, cols = img.shape
    new_img = np.zeros((rows, cols), dtype=np.uint8)
    center_row = rows // 2
    center_col = cols // 2
    for i in range(rows):
        for j in range(cols):
            new_i = int(i + a * (j - center_col))
            new_j = j
            if 0 <= new_i < rows and 0 <= new_j < cols:
                new_img[new_i, new_j] = img[i, j]
    return new_img

sheared_x_img = shearing_x(gray, 0.5)
plt.imshow(sheared_x_img, cmap='gray')
plt.title('Sheared Image in X-direction')
plt.show()
```

(E) Shearing in Y - direction

```python
#shearing in y-direction
#shearing wrt centre
#without using inbuilt functions
def shearing_y(img, a):
    rows, cols = img.shape
    new_img = np.zeros((rows, cols), dtype=np.uint8)
    center_row = rows // 2
    center_col = cols // 2
    for i in range(rows):
        for j in range(cols):
            new_i = i
            new_j = int(j + a * (i - center_row))
            if 0 <= new_i < rows and 0 <= new_j < cols:
                new_img[new_i, new_j] = img[i, j]
    return new_img

shear_y_img = shearing_y(gray, 0.5)
plt.imshow(shear_y_img, cmap='gray')
plt.title('Sheared Image in Y-direction')
plt.show()
```
